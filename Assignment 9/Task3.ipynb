```python
def add(n1,n2):
  """
  This function adds two numbers and returns the result

  Args:
  n1: First number
  n2: Second number

  Returns:
  n1+n2: Sum of two numbers
  False: If the n1 or n2 is invalid
  """
  if type(n1) == type(int()) and type(n2) == type(int()):
    return n1+n2
  else:
    print("Invalid input")
    return False
def subtract(n1,n2):
  """
  This function subtracts two numbers and returns the result

  Args:
  n1: First number
  n2: Second number

  Returns:
  n1-n2: Difference of two numbers
  False: If n1 or n2 is invalid
  """
  if type(n1) == type(int()) and type(n2) == type(int()):
    return n1-n2
  else:
    print("Invalid input")
    return False
def multiply(n1,n2):
  """
  This function multiplies two numbers and returns the result

  Args:
  n1: First number
  n2: Second number

  Returns:
  n1*n2: Product of two numbers
  False: If n1 or n2 is invalid
  """
  if type(n1) == type(int()) and type(n2) == type(int()):
    return n1*n2
  else:
    print("Invalid input")
    return False
def divide(n1,n2):
  """
  This function divides two numbers and returns the result

  Args:
  n1: First number
  n2: Second number

  Returns:
  n1/n2: Quotient of two numbers
  False: If n1 or n2 is invalid
  """
  if type(n1) == type(int()) and type(n2) == type(int()):
    if n2 == 0:
      print("Cannot divide by zero")
      return False
    return n1/n2
  else:
    print("Invalid input")
    return False
```
