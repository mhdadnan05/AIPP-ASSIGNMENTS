**Code:**

```python
def sum_of_even_and_odd_numbers(numbers_list):
  """
  This function takes a list (numbers_list) and returns the sum of even numbers, and sum of odd numbers in the list.

  Args:
  numbers_list: A list of numbers.

  Returns:
  even_nums_sum: Sum of all even numbers from the list.
  odd_nums_sum: Sum of all odd numbers from the list.
  False: If the list is empty or invalid elements in the list or given arguments other than list.
  """
  if type(numbers_list) == type(list()):
    even_nums_sum = 0
    odd_nums_sum = 0
    for number in numbers_list:
      if type(number) == type(int()):
        if number % 2 == 0:
          even_nums_sum += number
        else:
          odd_nums_sum += number
      else:
        return False
    return even_nums_sum, odd_nums_sum
  else:
    return False
```
